### Ejercicio 7 — JOIN desde Python
 
La tabla `mascotas` tiene el `dueno_id`, pero no el nombre del dueño. Para verlos juntos necesitás unir las dos tablas, igual que hacías en SQLBolt.
 
**Consigna:** Escribe una consulta con `JOIN` para traer el nombre de la mascota, su especie y el nombre de su dueño, todo en un solo DataFrame. No uses dos `pd.read_sql()` separados y luego los combines — hacelo en una sola consulta SQL.

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
df_mascotas_duenos = pd.read_sql(
    """SELECT mascotas.nombre AS mascota,
        mascotas.especie,
        duenos.nombre AS dueño
        FROM mascotas 
        JOIN duenos ON mascotas.dueno_id = duenos.id""",
    conexion
)
 
print(df_mascotas_duenos)
```

### Ejercicio 8 — Dos JOINs encadenados
 
Para ver el nombre del dueño junto con los datos de cada consulta médica, necesitas pasar por tres tablas: `consultas` → `mascotas` → `duenos`.
 
**Consigna:** Escribe una consulta que una las tres tablas y devuelva: nombre del dueño, nombre de la mascota, fecha de la consulta, motivo y costo. Ordená el resultado por fecha ascendente.

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
df_historial = pd.read_sql(
    """
        SELECT duenos.nombre AS dueno, mascotas.nombre AS mascota, consultas.fecha, consultas.costo, consultas.motivo
        FROM mascotas
        JOIN duenos ON mascotas.dueno_id = duenos.id 
        JOIN consultas ON consultas.mascota_id = mascotas.id
        ORDER BY consultas.fecha ASC

""",
    conexion
)
conexion.close()
print(df_historial)
```

### Ejercicio 9 — Investigación: alias en SQL

**Consigna:** Investiga cómo usar `AS` en SQL para renombrar columnas y tablas en una consulta. Después reescribí la consulta del Ejercicio 8 usando alias para:

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
df_historial = pd.read_sql(
       """
        SELECT duenos.nombre AS dueno, mascotas.nombre AS mascota, consultas.fecha AS fecha, consultas.costo AS costo, consultas.motivo AS motivo_consulta
        FROM mascotas
        JOIN duenos ON mascotas.dueno_id = duenos.id 
        JOIN consultas ON consultas.mascota_id = mascotas.id
        ORDER BY consultas.fecha ASC

""",
    conexion
)
 
print(df_historial.columns.tolist())
# Esperado: ['dueno', 'mascota', 'fecha', 'motivo', 'costo']
```
