```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conexion = sqlite3.connect("/workspaces/data-analysis-course/modulo-2-extraccion/clase-08/practica/ventas.db")

df = pd.read_sql(
    "SELECT * FROM productos", conexion
)

df["categoria"].value_counts()
df["categoria"].value_counts().plot(kind="bar")
plt.title("Categorías")
plt.xlabel("Categoría")
plt.ylabel("Cantidad")
plt.show()


sizes = [30, 25, 45]
labels = ['Apples', 'Bananas', 'Cherries']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal') # Ensures the pie chart is a perfect circle
plt.show()
```
